**Data Quality Assessment**

```python
missing = pd.DataFrame({
    'Missing_Count': df.isnull().sum(),
    'Missing_Persent': round((df.isnull().sum()/len(df))*100,2)
})

missing.sort_values(
    'Missing_Persent',
    ascending = False
)
```

```python
date_cols = [
    'Start_Date',
    'Primary_Completion_Date',
    'Completion_Date',
    'First_Posted',
    'Last_Update_Posted'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

```python
df['Trial_Duration_Days'] = (
    df['Completion_Date'] -
    df['Start_Date']
).dt.days
```

```python
df['Trial_Duration_Days'].describe()
```

```python
df[df['Trial_Duration_Days'] < 0][[
    'NCT_Number',
    'Study_Title',
    'Start_Date',
    'Completion_Date',
    'Trial_Duration_Days'
]]
```

```python
df.loc[
    df['Trial_Duration_Days'] < 0,
    'Trial_Duration_Days'
] = pd.NA
```

```python
df['Trial_Duration_Days'].describe()
```

```python
def enrollment_category(x):

    if pd.isna(x):
        return "Unknown"
    elif x < 100:
        return "Small"   
    elif x < 500:
        return "Medium"  
    else:
        return "Large"  
df['Enrollment_Category'] = df['Enrollment'].apply(enrollment_category)
```

```python
df['Enrollment_Category'].value_counts()
```

```python
df['Age_Group_Count'] = (
    df['Age']
    .fillna('')
    .str.split(',')
    .str.len()
)
```

```python
df[['Age','Age_Group_Count']].head()
```

```python
df.to_csv(
    "Clinical_trial_cleaned.csv",
    index=False
)

print("Clean datset saved succesfully!")
```
